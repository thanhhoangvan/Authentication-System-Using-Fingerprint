# Hệ xác thực sử dụng mật mã và dấu vân tay

```python
import ipywidgets as widgets
from IPython.display import display
import io
from PIL import Image
import matplotlib.pyplot as plt
from ipywidgets import FileUpload
from MASFinger import *

%matplotlib inline
```

## Mã hóa và lưu dấu vân tay

```python
upload = widgets.FileUpload(
    accept='.tif',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    description='Input Fingerprint',
    multiple=False  # True to accept multiple files upload else False
)

upload
```

```python
finger_path = './sample_inputs/' + list(upload.value.keys())[0]

widgets.Text(
    value=list(upload.value.keys())[0],
    placeholder='Type something',
    description='Fingerprint:',
    disabled=True
)

```

```python
for name, file_info in upload.value.items():
    img = Image.open(io.BytesIO(file_info['content']))

_ = plt.imshow(img)
```

```python
output = widgets.Output()

PIN1 = int(input("Vui lòng nhập mã PIN: "))
with open('code.txt', 'w') as code:
    code.write(str(PIN1))
encode_button = button = widgets.Button(description="Mã Hóa!")

display(encode_button, output)

def on_button_clicked(b):
    with output:
        if SaveFinger(finger_path):
            print("Đã mã hóa dấu vân tay")
        else:
            print("Mã hóa thất bại")
            
encode_button.on_click(on_button_clicked)
```
