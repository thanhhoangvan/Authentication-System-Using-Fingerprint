# Hệ xác thực sử dụng mật mã và dấu vân tay

```python
import ipywidgets as widgets
from IPython.display import display
import io
from PIL import Image
import matplotlib.pyplot as plt
from ipywidgets import FileUpload
from MASFinger import *

%matplotlib inline
```

## Xác thực vân tay

```python
upload2 = widgets.FileUpload(
    accept='.tif',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    description='Input Fingerprint',
    multiple=False  # True to accept multiple files upload else False
)

upload2
```

```python
finger_match = './sample_inputs/' + list(upload2.value.keys())[0]

widgets.Text(
    value=list(upload2.value.keys())[0],
    placeholder='Type something',
    description='Fingerprint:',
    disabled=True
)
```

```python
for name, file_info in upload2.value.items():
    img = Image.open(io.BytesIO(file_info['content']))

_ = plt.imshow(img)
```

```python
PIN2 = int(input("Vui lòng nhập mã PIN: "))
with open('code.txt', 'r') as code:
    PIN1 = int(code.read())
    
button2 = widgets.Button(description="Xác thực")
output2 = widgets.Output()

display(button2, output2)

def on_button_clicked_2(b):
    with output2:
        if PIN1 == PIN2:
            if MatchFinger(finger_match):
                print('Dấu vân tay khớp với vân tay đã lưu')
            else:
                print('Dấu vân tay không khớp hoặc sai mã PIN')
        else:
                print('Dấu vân tay không khớp hoặc sai mã PIN')

button2.on_click(on_button_clicked_2)
```
