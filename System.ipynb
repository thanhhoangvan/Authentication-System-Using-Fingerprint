# Hệ xác thực sử dụng mật mã và dấu vân tay

```python
import ipywidgets as widgets
from IPython.display import display
import io
from PIL import Image
import matplotlib.pyplot as plt
from ipywidgets import FileUpload

%matplotlib inline
```

## Chọn ảnh vân tay

```python
upload = widgets.FileUpload(
    accept='.tif',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    description='Input Fingerprint',
    multiple=False  # True to accept multiple files upload else False
)

upload
```

```python
widgets.Text(
    value=list(upload.value.keys())[0],
    placeholder='Type something',
    description='Fingerprint:',
    disabled=True
)
```

```python
for name, file_info in upload.value.items():
    img = Image.open(io.BytesIO(file_info['content']))

_ = plt.imshow(img)
```

## Mã hóa

```python
PIN = widgets.IntText(
    value=123456,
    description='PIN:',
    disabled=False
)

# print(PIN.value)


encode = widgets.ToggleButton(
    value=False,
    description='Mã hóa',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Description',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)

display(PIN, encode)
```

```python
img = plt.imread('output/101_1.png')
plt.title('Features Extraction')
plt.imshow(img)

print('Generate Mask from PIN: 10001101101011101000100011011010111010001110100011101000', '\n')
```
